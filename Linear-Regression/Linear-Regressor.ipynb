```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
def train_linear_regression_model(X, y):
    """
    Trains a linear regression model on the input features X and target y.
    Returns the trained linear regression model.
    """
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
    regressor = LinearRegression()
    regressor.fit(X_train, y_train)
    return regressor
```

```python
def plot_salary_vs_experience(X_train, y_train, model):
    """
    Plots the scatter plot of salary vs experience and the fitted line from the model.
    """
    plt.scatter(X_train, y_train, color='red')
    plt.plot(X_train, model.predict(X_train), color='blue')
    plt.title('Salary vs Experience (Training set)')
    plt.xlabel('Years of Experience')
    plt.ylabel('Salary')
    plt.show()
```

```python
def main():
    # Load the dataset
    dataset = pd.read_csv("C:/Users/canif/Desktop/video course/data analyst/1 - lesson/week1/Datasets-20230705/Salary_Data.csv")

    # Split the dataset into features and target variables
    features = dataset.iloc[:, :-1].values
    target = dataset.iloc[:, -1].values

    # Train the linear regression model
    linear_regression_model = train_linear_regression_model(features, target)

    # Predict and evaluate the model
    X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=1)
    y_pred = linear_regression_model.predict(X_test)
    train_score = linear_regression_model.score(X_train, y_train)

    # Format and print the coefficients and R-squared value
    coefficients = linear_regression_model.coef_
    intercept = linear_regression_model.intercept_
    formatted_coefficients = ', '.join([f'{c:.2f}' for c in coefficients])
    formatted_intercept = f'{intercept:.2f}'
    print("Coefficients (m):", formatted_coefficients)
    print("Intercept:", formatted_intercept)
    print("R-squared:", train_score)

    # Plot the results
    plot_salary_vs_experience(X_train, y_train, linear_regression_model)

if __name__ == "__main__":
    main()
```
