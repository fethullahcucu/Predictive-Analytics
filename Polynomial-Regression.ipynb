```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Load the dataset from the specified path
dataset_path = "C:/Users/canif/Desktop/video course/data analyst/1 - lesson/week1/Datasets-20230705/Position_Salaries.csv"
dataset = pd.read_csv(dataset_path)
```

```python
# Extract the feature (X) and target variable (y) from the dataset
X = dataset.iloc[:, 1:-1].values  # Feature: Position Level
y = dataset.iloc[:, -1].values    # Target Variable: Salary
```

```python
# Create a linear regression model and fit it to the data
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# Create a polynomial features object with degree 4 and transform the input features
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
```

```python
# Create another linear regression model using polynomial features and fit it to the data
lin_reg2 = LinearRegression()
lin_reg2.fit(X_poly, y)
```

```python
# Visualize the original data points and the fitted polynomial regression curve
plt.scatter(X, y, color="red", label="Actual Data")
plt.plot(X, lin_reg2.predict(poly_reg.fit_transform(X)), color="blue", label="Polynomial Regression")
plt.title("Polynomial Regression Model")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```python
# Make a prediction using the polynomial regression model for a specific position level (e.g., 6.5)
position_level_to_predict = 6.5
predicted_salary = lin_reg2.predict(poly_reg.transform([[position_level_to_predict]]))
print(f"Predicted Salary for Position Level {position_level_to_predict}: {predicted_salary[0]}")
```
